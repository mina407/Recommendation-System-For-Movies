#### Features Engineering

```python
## pre-processing geners column 
main_data['genres'][0]
```

###
* We need to extract the kind of each movies 

```python
## extracting the name of geners columns 
def get_name(row):
    names = []
    
    ## we indicate i for each list in each row
    for i in ast.literal_eval(row):
        names.append(i['name'])
    return names
```

```python
ast.literal_eval(main_data['genres'][0])
```

```python
## applying the func to get the kind of each movies 
main_data['genres'] = main_data['genres'].map(lambda x : get_name(x))
```

```python
main_data.head(3)
```

```python
## pre-processing cast columns
main_data['cast'][0]
```

###
* we need to extract the name of cast 

```python
def get_name_cast(obj):
    names = []
    counter = 0
    
    ## lets take only 3 name form each cast 
    for i in ast.literal_eval(obj):
        if counter !=3 :
            names.append(i['name'])
            counter += 1
        else :
            break
    return names
```

```python
main_data['cast'] = main_data['cast'].map(lambda x : get_name_cast(x))
```

```python
## pre-processing crew features 
ast.literal_eval(main_data['crew'][0])
```

### 
* We Need To Extract Director Name from Each Crew

```python
def get_director_name(obj):
    names = []
    
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            names.append(i['name'])
            
    return names
```

```python
main_data['crew'] = main_data['crew'].map(lambda x : get_director_name(x))
```

```python
ast.literal_eval(str(main_data['crew'][5]))
```

```python
## pre-processing keywords features 
ast.literal_eval(main_data['keywords'][0])
```

```python
def get_key_word(obj):
    names = []
    
    for i in ast.literal_eval(obj):
        names.append(i['name'])
    return names
```

```python
main_data['keywords'] = main_data['keywords'].map(lambda x : get_key_word(x))
```

```python
ast.literal_eval(str(main_data['keywords'][0]))
```

```python
main_data.head(5)
```

```python
## pre-processing overview
main_data['overview'][0]
```

```python
## Spliting each word 
main_data['overview'] = main_data['overview'].map(lambda x : x.split())
```

```python
main_data.head(3)
```

```python
main_data['genres'][0]
```

#### Removing The Space Between one Word

```python
for feature in ['genres','cast','crew','keywords']:
    
    ## x indicate each row and i indicate each word in each row
    main_data[feature] = main_data[feature].apply(lambda x : [i.replace(" " , "") for i in x])
```

```python
main_data.head()
```

```python
## Combine the features
main_data['comb'] = main_data['genres'] + main_data['cast'] + main_data['keywords'] + main_data['crew'] + main_data['overview']
```

```python
main_data = main_data[['movie_id' , 'title' , 'comb']]
main_data.head()
```

```python
## combine each element in the list
main_data['comb'] = main_data['comb'].apply(lambda x : " ".join(x))
```

```python
## Lower case 
main_data['comb'] = main_data['comb'].apply(lambda x : x.lower() )
```

```python
## Split the data 
X = main_data['comb']
```

```python
## transform the data 
vector = TfidfVectorizer()

X = vector.fit_transform(X)
```

```python
## find the similarity for movies 
similarity = cosine_similarity(X)
```

```python
sorted(list(enumerate(similarity[0])) , key=lambda x : x[1] , reverse=True)[1:6]
```

```python
def recommendation(movie):
    index = main_data[main_data['title'] == movie].index[0]
    
    movie_list = sorted(list(enumerate(similarity[index])) , key=lambda x : x[1] , reverse=True)[1:20]
    for i in movie_list:
            print(main_data.iloc[i[0]].title)
```

```python
recommendation("Pirates of the Caribbean: At World's End")
```
